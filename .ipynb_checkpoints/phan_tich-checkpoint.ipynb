```python
import pandas as pd

df = pd.read_csv(r'KHẢO SÁT NHÓM 5.csv')
df[['Cho chúng mình biết bạn đang là sinh viên khóa bao nhiêu ?','Mẫu máy của bạn là gì ?']]
```

```python

df2=df.rename({'Cho chúng mình biết bạn đang là sinh viên khóa bao nhiêu ?': 'Year','Chuyên ngành của bạn là gì ? ':'Major','Giới tính của bạn':'Sex','Bạn đang sử dụng laptop hãng nào ?':'Brand','Mẫu máy của bạn là gì ?':'Model','Mức độ hài lòng về chiếc laptop của bạn':'Overall_satisfaction',
          'Đánh giá sơ bộ laptop của bạn [Chất lượng hoàn thiện]':'Build_quality','Đánh giá sơ bộ laptop của bạn [Thiết kế]':'Design','Đánh giá sơ bộ laptop của bạn [Thời lượng pin]':'Battery_life','Thời lượng sử dụng laptop của bạn mỗi ngày':'Time_usage','Mục đích sử dụng laptop của bạn là gì\n(Bạn có thể chọn nhiều hơn 1 câu trả lời)':'Purpose',
          'Laptop có giúp ích trong việc học và làm của bạn không':'Useful','Bạn có sử dụng laptop khi lên lớp không ?':'Class_use','Bạn có thường bị mất tập trung khi sử dụng laptop cho việc học không ?':'Distraction_rate','Bạn dành bao nhiêu tiền để mua laptop (đơn vị Triệu đồng)':'Bud',
          'Nếu được chọn lại, bạn sẽ chọn mẫu máy laptop nào? Tại sao?':'Rechoose','Thời lượng sử dụng laptop của bạn trong việc học trong 1 ngày':'Time_study','Trên thang điểm 1-5 hãy đánh giá khả năng sử dụng máy tính của bạn (1: Mới sử dụng, 5: Thành thạo)':'Skill','Bạn đã từng sử dụng mẹo để tăng tốc độ xử lý của máy tính chưa?':'Trick'},
          axis=1)
df2['Major']=df2['Major'].str.lower()

map1={'Bình thường':3,'Khá hài lòng':4,'Rất hài lòng':5,'Không hài lòng':2,'Thất vọng':1}
map2={'Đủ':4,'Ngoài mong đợi':5,'Bình thường':3}
map3={'Ít khi':2,'Thình thoảng':4,'Bình thường':3,'Thường xuyên':5,'Không':1}
map_price={'< 10':7.5,'10-15':12.5,'15-20':17.5,'20-25':22.5,'25+':30}
map_time={'2 - 4 tiếng':3,'30 phút - 1 tiếng':0.75,'1- 2 tiếng':1.5,'4 tiếng +':5,'ít hơn 30 phút':0.25}
df2['Design']=df2['Design'].map(map1)
df2['Build_quaily']=df2['Build_quality'].map(map1)
df2['Battery_life']=df2['Battery_life'].map(map1)
df2['Overall_satisfaction']=df2['Overall_satisfaction'].map(map1)
df2['Useful']=df2['Useful'].map(map2)
df2['Class_use']=df2['Class_use'].map(map3)
df2['Time_study']=df2['Time_study'].map(map_time)
df2['Time_usage']=df2['Time_usage'].map(map_time)
#df2['Distraction_rate']=df['Distraction_rate'].map(map3)
#df2['Bud']=df['Bud'].map(map_price)
df2['Distraction_rate']=df['Distraction_rate'].map(map3)
df2
```

```python
df2['Count_brand']=1
df3=df2.groupby(['Model'])[['Brand','Useful','Class_use','Overall_satisfaction','Time_study','Time_usage']].mean()
df3['Count']=df2.groupby(['Model']).sum()['Count_brand']
df3['Learn_rate']=(df3['Time_study']/df3['Time_usage'])*100
df3.sort_values(by=['Count'],ascending=False)
```

```python
df3=df2.groupby(['Brand'])[['Brand','Overall_satisfaction','Build_quality','Design','Battery_life']].mean()
df3.sort_values(by=['Count'],ascending=False)
```
